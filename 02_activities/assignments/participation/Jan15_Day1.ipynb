```python
# Making a basic figure with matplotlib
# Generate some sample data for plotting (x,y) coordinates
# Generate a random seed so we can reproduce the figure
np.random.seed(613)

# Generate 50 random x integers from 0-49
x = np.arange(50)

# Generate 50 random y integers from 0-100 
y = np.random.randint(0, 100,50)
```

```python
# Make a basic scatterplot
# Use a matplotlib.pyplot function called plt.subplots; pyplot is mainly intended for interactive plots

# Specify the width and height of the subplot (axes). It creates two objects: 1. the figure objects (i.e., the entire canvas), and 2. the axes (the plotting area where we draw). 
# This is how matplotlib encodes the figure. 
fig, ax = plt.subplots(figsize=(5, 3))

# Use scatter function to generate a scatterplot for x,y coordinates
ax.scatter(x,y)
```

```python
# Modify the graph
# Keep the figure and axis definition the same; therefore, only edit our axis object

# Make a barplot (rather than a scatterplot) usign the same setup
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# Make a line plot instead
# Use the same figure object 
fig, ax = plt.subplots(figsize=(5, 3))
# Edit the axes object, using the plot function
ax.plot(x,y)
```

```python
# Make a histogram
# Note: A histogram only shows the y values' frequencies, so only need to call the y integer values
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# To add bin boundaries
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y, edgecolor='black')
```

```python
# Adding axis lables and titles
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# Add to the last plot we made
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')

# Adjust the specing between the x and y axis to avoid overlapping between subplots and margins
fig.tight_layout()
```

```python
# Modiying labels and titles
# Use font dictionaries('fontdict') 

# First, make a dictionary with the properties we want for our labels
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}
```

```python
# Add fontdict arguments to our previous lineplot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()
```

```python
# Customize the plot appearance by setting up the shape and colour of of each data point

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,
           y, 
            marker='*', # NEW ADDITIONS marker = '*'
            color = "indigo")  # NEW ADDITIONS color = "indigo")
fig.show()
```

```python
# Alternative code using "plt.show()"
fig,ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, marker='*', color="indigo")
plt.show()
```

```python
# Adjust the style and width of the line connecting data points
fig, ax = plt.subplots(figsize=(5, 3))
# Use "ax.plot" function to draw a line plot instead
ax.plot(x,
           y, 
            marker='*', 
            color = "indigo",
            linestyle = '--', #NEW ADDITONS 
            linewidth = 2) # NEW ADDITIONS
fig.show()
```

```python
# Modifying data points: Colour
# We can modify colour using named colors (See link directly above to see hex codes)
# Use hex codes

fig, ax = plt.subplots(figsize=(5, 3))

# New addition: color
ax.plot(x,y,marker='*', color = "#7425b9", linestyle = '--', linewidth = 2)
fig.show()
```

```python
# Further customize plot - Note plotting line colour is still indigo; if blanked out, default is light blue
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y,marker='*', 
        markersize = 12, # NEW ADDITIONS
        #color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', # NEW ADDITIONS 
        markerfacecolor = '#000000' ) # NEW ADDITIONS
```

```python
# Grid lines
# Add grid lines to our axis object using .grid() function
fig, ax = plt.subplots(figsize=(5, 3))
# Insert y axis grid line
ax.grid(axis='y')
ax.plot(x,y,marker='*', 
        markersize = 12, # NEW ADDITIONS
        #color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', # NEW ADDITIONS 
        markerfacecolor = '#000000' ) # NEW ADDITIONS
```

```python
# Modify the axes appearance (colour, width, linestyle) or our grid lines
fig, ax = plt.subplots(figsize=(5, 3))
# Insert y axis grid line
ax.grid(axis='y', color = "blue", linewidth = 2, linestyle = '-.')
ax.plot(x,y,marker='*', 
        markersize = 12, # NEW ADDITIONS
        #color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2, 
        markeredgecolor = '#fa9359', # NEW ADDITIONS 
        markerfacecolor = '#000000' ) # NEW ADDITIONS
```

A basic lollipop plot can be created using the stem() function of matplotlib. This function takes x axis and y axis values as an argument. x values are optional; if you do not provide x values, it will automatically assign x positions.

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
 
# create data
x=range(1,41)
values=np.random.uniform(size=40)
 
# stem function
plt.stem(x, values)
plt.ylim(0, 1.2)
plt.show()
 
# stem function: If x is not provided, a sequence of numbers is created by python:
plt.stem(values)
plt.show()
```

Aesthetically, this lollipop chart is simple and attractive to view. However, substantively it does not highlight any meaningful data category comparisions. In contrast, a histogram would be better since it would produce meaningful binning of the y values by frequency. Rather, it only show the apparent variation among the randomly-generated data points. Perceptually, a lollipop chart presentation of this data set is overwhleming and does serve to provide any meaningful distinctions in the data set.

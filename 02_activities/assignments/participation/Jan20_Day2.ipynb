## Choosing dataset types: tables, networks, fields, geometry (also: multidimensional table, tree)
- Determine attribute comparisons made--> tables
	- more two indexing dimensions --> multidimensional tables
- Relationships between items --> networks
	- hierarchy of items within the network --> trees
		- structures and levels matter
- Attributes over a range of values --> fields
- Objects have unique shape or spatial location --> geometry

# Effective visualization
- Depends on purpose, audience and medium (see above)
- Spaces: known--> consideration --> proposal --> selected (see slide page 29)
- Need to have a large space of possible solutions available to explore/consider
- Not just about execution
- Different purposes --> different results

# Visualization perceptions
- Use human cognition, including for example Gestalt principles:
	- proximity
	- similarity
	- continuity
	- closure
	- enclosure
	- connection
- Consider cognitive load as well --> how much effort is required to absorb new information. Levels:
	- intrinsic 
	- germane (audience's familiarity with the information)
	- extraneous (presentation complexity - to perceive graph structure)
- We can only control how we present the information:
	- familiarity of chart type
	- relational vs. absolute values 
	- conciseness of the composition
	- explanatory vs. exploratory
- Working memory is limited --> better to show comparisons side-by-side
	- 3D increases cognitive load: use cautiously
	- Also makes category comparisons more challenging
	- But some relationships can only be portrayed in 3Ds (e.g., wind tunnel)

# Provenance rhetoric
- cite sources for reproducibility and trust
- by including data source within the graphic, signal "transparency and trustworthiness"
- provenance rhetoric increases believability and the data visualization's persuasiveness

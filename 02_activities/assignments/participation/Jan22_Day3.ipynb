```python
# Make more modifications to our earlier plots using Matplotlib
# Legends
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Generate sample data

np.random.seed(613)
# x random values 0-49
x = np.arange(50)
# y1 & y2 random values 0-99, generate 50 of them
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# Line plot with both y variables on the *same* axis

fig, ax = plt.subplots(figsize=(5, 3)) # specify plot width & height
ax.plot(x,y1)
ax.plot(x,y2)
# show the entire figure canvas
fig.show()
```

```python
# Adding a legend
fig, ax = plt.subplots(figsize=(5, 3))

# Add a label argument to each plot object
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )

# Call a legend argument and specify the legend location
ax.legend(loc='lower right')
fig.show()
```

```python
# Adding a legend in centre of the plot
fig, ax = plt.subplots(figsize=(5, 3))

# Add a label argument to each plot object
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )

# Call a legend argument and specify the legend location
ax.legend(loc='center')
fig.show()
```

```python
# Add arguments to the legend object to change the legend elements

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = True, #add a frame around the legend
fontsize = 12, #change legend font size
ncol = 2, #specify number of columns for the legend (here: each y line has a separate column)
shadow = True )
fig.show()
```

frameon = False --> for transparency

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='lower right',
frameon = False, #add a frame around the legend
fontsize = 12, #change legend font size
ncol = 2, #specify number of columns for the legend (here: each y line has a separate column)
shadow = True )
fig.show()
```

```python
# Move legend location outside of plot area

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc="upper left",bbox_to_anchor =(1, 1)) # Anchor location on the figure using "bbox_to_anchor" - here, top-right of plotting area using "1,1"
fig.show()
```

```python
# Text() function annotations

# Generate a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# Use text function to specify plot coordinates and the text content
ax.text(10, 95, "This value is important!") # use the text()funciton to specify coordinates and text
fig.show()
```

```python
# Add arguments to modify existing text function annotations using data point coordinates
# ha == horizontal alignment
# Note: Also works without specifying fig.show()

# Generate the scatter plot again
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
    ha='center', ##alignment of text
    color = 'red', ##modify font colour
    size = 20) ##modify font size
```

```python
# Change text font

# Generate the scatter plot again
fig,ax=plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label='Person 1')
ax.scatter(x,y2,label="Person 2")
ax.legend(loc='lower right',prop={"family":"serif","size":20})
```

```python
# Position text annotations
# Fix annotation to a specific axes or figure location
# Method 1 Use transforms

# Generate empty axes sets
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
# Position the annotation to the data 
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)
```

```python
# Position annotation to data axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)
```

```python
# Both positions 
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
# Position the annotation to the data 
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes)
```

```python
# Position annotation to the figure (referenceing axis/figure dimensions)
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# Distinguish all three modifications: 
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])
ax.text(1, 5, ". Data: (1, 5)", transform=ax.transData ) # data point alignment
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes ) # align with plot axis (as a % of the entire axis)
ax.text(0.5, 0.1, ". Figure: (0.5, 0.1)", transform=fig.transFigure ) # align with reference to the proportion of the entire figure canvas  --> bigger than the plot area
```

```python
# Method 2 Use Arrows to specify a data point
# Use the annotate() function to annotate with arrows and text (e.g. to focus at aspecific point)
# Add arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), # Specify location to point at, xy=() & text location, xytext=()
    arrowprops=dict(facecolor='black')) # Arrow color
fig.show()
```

```python
# Modify existing arrows
# Use the arrowpros argument to change arrow style and colour
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.annotate("This is important!",
    xy=(10, 95), xytext=(20, 94),
    arrowprops = dict(arrowstyle = "wedge",
                        color = "hotpink"))
fig.show()
```

```python
# Axes labels
# Remove tick marks and lables with locator function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
# Remove the y axis label and tick marks using the "Nulllocator" argument with the "set_major_locator" function
ax.yaxis.set_major_locator(plt.NullLocator())
```

```python
# Remove labels only with formatter function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
# Modify the labels *only* (not the tick marks), using the "NullFormator" argument
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

```python
# Define maximum number of ticks to limit tick marks with MaxNLocator argument
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
# Specify number of tick marks on the axes
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# Set tick marks interval with MultipleLocator function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
# Set interval length to every 5 units
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
# Rotate axis labels and horizontal alignment with xtick function
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
# right side of the text attaches/closest to the tick mark
plt.xticks(rotation=45, ha='right')
```

```python
# Modify fontsize and font
font1 = {' family':'serif','color':'indigo'} # Select fontdict and font colour
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
plt.xlabel('Shiny New X Axis!', fontsize = 18, fontdict = font1) # Use xlabel function to label x axis with the selected font type and font colour
```

```python
# Selecting plot styles
# Use style.use function to select a particular plot style
# Use "fivethirtyeight" plot style
plt.style.use('fivethirtyeight)
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

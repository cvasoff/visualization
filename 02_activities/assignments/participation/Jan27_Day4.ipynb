```python
# Load Python libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Generate sample data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# Define multiple axes and their relative positions
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
```

```python
# Add figure subplots
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
fig.show()
```

```python
# Within a figure: Subplots grid arrangement using plt.subplot_mosaic()
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
    ['ax2', 'ax3']],
    figsize=(7, 4))
```

```python
# Add subplots' data
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
    ['ax2', 'ax3']],
    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
# Layouts
# Add an x axis titles to the figure above
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```

```python
# Two layout types to consider
# 1. Tight layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
fig.tight_layout()
```

```python
# 2. Constrained layout
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4),
layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label', fontsize=18)
someaxes["ax2"].set_xlabel('Another Label', fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize=18)
```

```python
# Multiple viz on one axes object: Call multiple plots on the same axes
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

```python
# Add error information
# 1. Calculate the data standard deviation
y2_sd = np.std(y2)
# Create a line plot for the SD by x varaible values
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
```

```python
# Use existing sample data plot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
# Add error information (i.e., an error line) with the errorbar function
ax.errorbar(x, # x values
y2, #y values
yerr = y2_sd, # Specifies vertical error bars plot
fmt = "none") # Plot the error information only, not the actual data points
```

```python
# Customize errorbar appearance
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")
ax.errorbar(x,
y2,
yerr = y2_sd,
fmt = "none",
ecolor= "indigo",
elinewidth= 4,
capsize = 6,
capthick= 4 )
```

```python
# Specify errorbar intervals to avoid displaying bars for every data point
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
color = "red")
ax.errorbar(x,
y2,
yerr = y2_sd,
fmt = "none",
ecolor = "indigo",
elinewidth = 4,
capsize = 6,
capthick = 4,
errorevery= 2 )
```

```python
# Adding images with images from the internet
# Load the relevant Python libraries
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
# Retrieve internet images
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# Generate a basic line plot coded earlier
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
# Create a figure container for the image by overlaying a new figure (i.e., "ax") axis 
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES)
                        0.11, # y coordinate (ON FIGURE, NOT AXES)
                        0.15, # image width
                        0.35]) # image height
```

```python
# Add image to the plot with imshow function
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2,
        color = "red")
ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])
ax_image.imshow(image)
ax_image.axis("off")
```
